```python
print("=" * 55)
print("STUDENT PERFORMANCE EXPLORER ")
print("=" * 55)
```

```python
print("\n DATASET OVERVIEW")
print("-" * 35)
print(f"  Total Students: {len(df)}")
print(f"  Total Columns: {len(df.columns)}")
print(f"  Unique Departments: {df['department'].nunique()}")
```

```python
print("\nDEPARTMENT-WISE STUDENT COUNT")
print("-" * 35)
dept_counts = df["department"].value_counts()
for dept, count in dept_counts.items():
    print(f"  {dept:<15} : {count} students")
```

```python
print("\nSCORE ANALYSIS")
print("-" * 35)
print(f"  Highest Math Score: {df['math_score'].max()}")
print(f"  Lowest Math Score: {df['math_score'].min()}")
print(f"  Avg Math Score: {df['math_score'].mean():.2f}")
print(f"  Avg Programming Score: {df['programming_score'].mean():.2f}")
print(f"  Avg Attendance %: {df['attendance_percentage'].mean():.2f}%")
```

```python
print("\nGENDER-WISE AVG TOTAL SCORE")
print("-" * 35)
gender_avg = df.groupby("gender")["total_score"].mean()
for gender, avg in gender_avg.items():
    print(f"  {gender:<10} : {avg:.2f}")
```

```python
print("\nTOP 5 PERFORMERS")
print("-" * 45)
print(f"  {'Rank':<5} {'Name':<12} {'Department':<14} {'Total Score'}")
print("  " + "-" * 42)
top5 = df.sort_values("total_score", ascending=False).head(5).reset_index(drop=True)
for i, row in top5.iterrows():
    print(f"  {i+1:<5} {row['name']:<12} {row['department']:<14} {row['total_score']}")
```

```python
print("\nSTUDENTS WITH ATTENDANCE BELOW 75%")
print("-" * 45)
low_att = df[df["attendance_percentage"] < 75]

if low_att.empty:
    print("All students have attendance above 75%!")
else:
    print(f"  {'Name':<12} {'Department':<14} {'Attendance %'}")
    print("  " + "-" * 38)
    for _, row in low_att.iterrows():
        print(f"  {row['name']:<12} {row['department']:<14} {row['attendance_percentage']}%")
```

```python
min_prog_idx = df["programming_score"].idxmin()
min_prog_student = df.loc[min_prog_idx, "name"]
min_prog_score   = df.loc[min_prog_idx, "programming_score"]
print(f"Q3. Lowest programming score             : {min_prog_student} ({min_prog_score})")
```

```python
above_80_math = len(df[df["math_score"] > 80])
print(f"Q4. Students who scored above 80 in math : {above_80_math}")
```

```python
avg_total_high_att = df[df["attendance_percentage"] > 90]["total_score"].mean()
print(f"Q5. Avg total score (attendance > 90%)   : {avg_total_high_att:.2f}")

print("\n" + "=" * 55)
print("ANALYSIS COMPLETE")
print("=" * 55)
```
